# Задание 2:
Построить графики всех элементарных функций с данного сайта: http://mathprofi.ru/grafiki_i_svoistva_funkcij.html

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

def showPlot(x_start, x_max, step, callback, title='Function'):
    x = np.arange(x_start, x_max, step)
    y = callback(x)

    plt.plot(x, y)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title(title)
    plt.grid(True)
    plt.show()
```

```python
showPlot(-10, 10, 1, lambda x: -1*x+1, title='Линейная функция')
```

```python
showPlot(-10, 10, 1, lambda x: x*x, title='Парабола')
```

```python
showPlot(-10, 10, 1, lambda x: x*x*x, title='Кубическая парабола')
```

```python
showPlot(0, 10, 0.1, lambda x: np.sqrt(x), title='Квадратный корень')
```

```python
showPlot(.1, 10, 0.1, lambda x: 1/x, title='Гипербола')
showPlot(-10, -.1, 0.1, lambda x: 1/x, title='Гипербола')
```

```python
showPlot(-10, 10, 0.1, lambda x: np.exp(x), title='Показательная функция')
```

```python
showPlot(.1, 10, 0.1, lambda x: np.log(x), title='Логарифмическая функция')
```

```python
showPlot(-10, 10, 0.1, lambda x: np.sin(x), title='Sin')
```

```python
showPlot(-10, 10, 0.1, lambda x: np.cos(x), title='Cos')
```

```python
showPlot(.5, np.pi-.5, 0.01, lambda x: np.cos(x)/np.sin(x), title='Ctg')
```
