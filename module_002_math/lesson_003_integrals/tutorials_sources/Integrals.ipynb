# Интегралы


**Интеграл** - это "действие" обратное производной

$v(t)=t(8-t)$

$s(t)=\int v(t) dt$



```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from sympy import *

def showPlot(x_start, x_max, step, callback, title='Function', x_title='X', y_title='Y'):
    x = np.arange(x_start, x_max, step)
    y = callback(x)

    plt.plot(x, y)
    plt.xlabel(x_title)
    plt.ylabel(y_title)
    plt.title(title)
    plt.grid(True)
    plt.show()
    
def value_of_expr(x_vector, symbol, expr):
    y = np.empty_like(x_vector)
    for index, element  in enumerate(x_vector):
        y[index] = (expr.subs(symbol, element))
    return y

init_printing()
```

```python
t, v = symbols('t v')

vt = t*(8-t)
showPlot(0, 8, 0.01, lambda i: value_of_expr(i, t, vt), title='Velocity', x_title='time', y_title='velocity')
vt
```

```python
distance = integrate(vt, t)
integral
```

```python
showPlot(0, 8, 0.01, lambda i: value_of_expr(i, t, distance), title='Distance', x_title='time', y_title='distance')
```
